```python
# Importing and setting up
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # additional plotting tools
from sosp import simple
from itertools import product
```

```python
# Verb choice: settle on a noun-verb link -> number marking
centers = np.array([[1, 0], [0, 1.]]) # N1-V, N2-V
gamma = 0.2  # 0.4 = default, but this is what I used for the diss. simulations
```

```python
# Trying starting w/ 0.5 b/c very low h_i --> failure to build a parse
h0seq = np.round(np.arange(0.05, 1.05, 0.05), 2)
h1seq = np.round(np.arange(0.05, 1.05, 0.05), 2)
cond_names = []
for i, j in product(h0seq, h1seq):
    cond_names.append('h0_{}_h1_{}'.format(i, j))
```

```python
cond_harmonies = np.array(np.meshgrid(h0seq, h1seq)).T.reshape(-1, 2)
print(cond_harmonies.shape)
#cond_harmonies
```

```python
xinit = np.zeros(centers.shape[1])
sys = simple.SimpleModel(corpus_filename=None, centers=centers, local_harmonies=np.array([1., 1.]),
                         stopping_crit='cheb_stop')
sys.set_gamma(gamma)
sys.set_noise_mag(0.0005)  # What was used in the diss.
sys.locate_attrs()
```

```python
data = sys.run_multiple_conditions(n_runs=200, conditions=cond_harmonies)
```

```python
means = data.groupby('Condition').agg(['mean'])
means.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X, Y = np.meshgrid(h0seq, h1seq)
Z = means.Time.values.reshape(len(h0seq), len(h1seq))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
plt.xlabel('$h_0$')
plt.ylabel('$h_1$')
ax.zaxis.set_label('Mean RT')
#ax.view_init(azim=-135)  # -135 works well
plt.show()
```

```python
f1 = lambda x: 0.8*x
f2 = lambda y: 5./4*x
x = np.arange(0.1, 0.9, 0.1)

plt.figure(figsize=[8, 8])
plt.imshow(Z, extent=[0.05, 1.0, 0.05, 1.0], origin='lower', cmap=plt.cm.binary)
plt.colorbar().set_label('Mean settling time')
plt.axis(aspect='image')
plt.plot(x, f1(x), 'r')
plt.plot(x, f2(x), 'r')
plt.xlabel('$h_0$')
plt.ylabel('$h_1$')
plt.show()
```
